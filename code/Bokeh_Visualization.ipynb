```python
def group(x):
    if x > 60:
        return 4
    elif x > 35:
        return 3
    elif x > 18:
        return 2
    else:
        return 1
```

```python
df_age["Age group"] = df_age["Age"].apply(group)
```

```python
df_age
```

```python
category_list = list(df_age.columns.values)[12:47]
```

```python
#list(df_age.columns.values)
```

```python
group_1 = list()
for category in category_list:
    group_1.append(df_age.loc[df_age['Age group'] == 1,category].sum())
    
```

```python
#group_1
```

```python
group_2 = list()
for category in category_list:
    group_2.append(df_age.loc[df_age['Age group'] == 2,category].sum())
```

```python
group_3 = list()
for category in category_list:
    group_3.append(df_age.loc[df_age['Age group'] == 3,category].sum())
```

```python
group_4 = list()
for category in category_list:
    group_4.append(df_age.loc[df_age['Age group'] == 4,category].sum())
```

```python
total = list()
for category in category_list:
    total.append(df_age[category].sum())
```

```python
#category_list
```

```python
data = pd.DataFrame({'Category': category_list, 'Total': total, 'Age <= 18': group_1, '18 < Age <= 35': group_2, '35 < Age <= 60': group_3, 'Age > 60': group_4})
```

```python
from bokeh.charts import Bar, output_file, show
from bokeh.charts.attributes import cat, color
from bokeh.charts.operations import blend
#from bokeh.charts.utils import df_from_json
#from bokeh.sampledata.olympics2014 import data
```

```python
data = data[data['Total'] > 20]
len(data)
```

```python
data = data.sort_values(by="Total", ascending=False)
data
```

```python
from bokeh.charts import BoxPlot, output_file, show
from bokeh.layouts import row
from bokeh.layouts import column
```

```python
bar = Bar(data,
          values=blend('Age <= 18', '18 < Age <= 35', '35 < Age <= 60', 'Age > 60', name='Age Groups', labels_name='group'),
          label=cat(columns='Category', sort=False),
          stack=cat(columns='group', sort=False),
          color=color(columns='group', palette=['#FDE724', '#79D151', '#22A784','#29788E'],
                      sort=False),
          title="User Age Groups by Category",
          legend='top_right',
          tooltips=[('group', '@group'), ('category', '@Category')],
            plot_width = 600, plot_height = 600)

box2 = BoxPlot(df, values='Price', label='Category', color='Category',
               title="Book Price Box Plot by Category", plot_width=600)

output_file('/Users/siliwang/Desktop/bokehplots.html')
show(column(bar, box2))

```
