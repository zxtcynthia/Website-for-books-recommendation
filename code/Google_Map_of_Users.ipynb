# Google Map of Users 

```python
import pandas as pd
file='f:\Columbia\S4572_Data analytics for OR\Project\map long-lat2.csv'
```

```python
maps=pd.read_csv(file)
```

```python
maps=maps.values.flatten()
maps
```

```python
len(maps)
a=[]
b=[]
```

```python
for i in range(len(maps)):
    maps[i]=maps[i].replace("'",'')[1:-2]
    maps[i]=maps[i].split(',')
    a.append(float(maps[i][0]))
    b.append(float(maps[i][1]))
```

```python
a=pd.DataFrame(a)
b=pd.DataFrame(b)
```

```python
frames = [a,b]
result = pd.concat(frames,axis=1)
result.columns = ['latitude', 'longitude']
result
```

```python
result=result[result['latitude']<49.4]
result=result[result['latitude']>24.5]
result=result[result['longitude']<-66.9]
result=result[result['longitude']>-124.7]
result
```

```python
l=[]
for i in range(result.shape[0]):
    dict={}
    dict['lat']=result.iloc[i,0]
    dict['lng']=result.iloc[i,1]
    l.append(dict)
    
```

```python
l
```

```python
with open("file.txt", 'w') as f:
    for s in l:
        f.write(str(s) + '\n')
```
